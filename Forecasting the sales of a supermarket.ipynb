```python
df['Order Month'] = pd.to_datetime(df['Order Date'])
df['Ship Month'] = pd.to_datetime(df['Ship Date']).dt.strftime('%B')
df['Order Month']=pd.to_datetime(df['Order Month']).dt.strftime('%B')
df['Order Month']
```

```python
df.info()
```

```python
Segment_percent=df['Segment'].value_counts(normalize= True)*100
Segment_percent
```

# Handle City Column

```python
city_count=df.City.value_counts()
city_count_under50 = city_count[city_count<50]
```

```python
def handle_city(value):
    if value in city_count_under50:
        return ('Others')
    else :
        return(value)


df['City'] = df['City'].apply(handle_city)
df.City.value_counts()
```

# Handle Product Name Column

```python
df['Product Name'].value_counts()
```

```python
Product_Name_count=df['Product Name'].value_counts()
Product_Name_5times= Product_Name_count[Product_Name_count<5]
```

```python
def handle_Product_Name(value):
    if value in Product_Name_5times:
        return("Others")
    else :
        return(value)


df['Product Name']=df['Product Name'].apply(handle_Product_Name)
```

```python
df['Product Name'].unique()
```

# Handle Category Column

```python
Category_per_month= df.groupby('Order Month')[['Category']].value_counts().reset_index().set_index ('Order Month')
Category_per_month
```
